## 1. Deep Learning

Deep Learning (DL) is an approach to Machine Learning, where inputs are still used to predict outputs.

It works through networks of artificial neurons, which progressivelly extract higher level features from data.

### 1.1. Artificial Neurons

Artificial Neurons try to mimic the human brain, by combining input information and applying an activation function, to fire an output.

</div>

* $x_i$ are the inputs
* $w_i$ are the weights (trainable)
* $b$ is the bias term (can also be trained)
* $f$ is a non-linear activation function to compute the neuron’s output.

### 1.2. Deep Neural Networks

Deep Neural Networks (DNN) combine multiple artificial neurons throughout multiple layers before making a prediction.

* __Input Layer__: Receives the input data ths passes the information (features) to the hidden layer.

* __Hidden Layer__: Performs all computations on the features entered through the input layer and transfers the result to the output layer. 

* __Output Layer__: The final layer of the network that brings the information learned through the hidden layer and delivers the final value as a result. 

### 1.3. Types of Deep Neural Networks

In the DL world, there are multiple approaches to train DNN, which explore and take advantage of different specifications of data.

##### 1.3.1. Multilayer Perceptrons (MLPs)

Multilayer Perceptrons are the most traditional type of DNN, composed of multiple fully connected layers of neurons, where each neuron receives as the input all available data.

##### 1.3.2. Convolutional Neural Networks (CNNs)

This type of networks has the ability to explore spacial dependencies between input data, useful in structured data such as images. CNNs capture these relationships through the application of relevant filters. 

The computational cost of CNNs is also lower has it is possible to reuse weights (filters slide through the structure). These convolutional layers perform feature extraction.

When input data is composed of a sequence of dependent values (e.g. time series), where some value depends on the previous (and/or) next values, it is useful to explore this relationship.

In such cases, Recurrent Neural Networks aid the learning process with memory units, storing values of previous estimations.

### 1.4. Activation Functions

An activation function works as a gate on each neuron, to decide if it should be activated or not. Just like in humans' neurons, information is only transmitted (neurons fire an electric signal) if its input signals are relevant to be forwarded.

Activation functions add non-linearity to neural networks, avoiding that neuron only apply linear transformations on the inputs, from weights and biases. Without activation functions, independently of how deep a network becomes, all layers would behave in the same way because the composition of two linear functions is a linear function itself.

### 1.5. Loss Functions

An objective function is the method used to evaluate a candidate solution for the problem in hands. This function should either be minimized or maximized during training.

When the objective is to minimize the error of the problem, the objective function is often referred to as a cost function or a loss function.

During training, the loss function is computed iteratively to verify the difference between the network's predictions and the real expected values.

* __Regression__ - Mean Squared Error 

$$\text{MSE} =  \frac{1}{N}  \sum_{i=1}^N  ( y_i - \hat{y}_i )^2$$

* __Classification__ - Cross Entropy/Log Loss

$$\text{Binary Cross Entropy} = - \frac{1}{N} \sum_{i=1}^N   y_i*\log{(\hat{y}_i)} + (1 - y_i)\log{(1 -\hat{y}_i)}$$

$$\text{Categorical Cross Entropy} = - \frac{1}{N} \sum_{i=1}^N  \sum_{j=1}^K y_{ij}*\log{(\hat{y}_{ij})}$$

N is the number of evaluated samples and K the number of classes (categories).


### 1.6. How do Neural Networks Learn?

As loss functions measure the error of the network, the closer they are to 0 the better. Therefore, the objective of the training process is to __minimize the loss function__, which can be achieved through a apecific combination of all network's parameteres (weights and biases).

One could try to guess the values all weights and biases should take, but DNN are highly complex, with millions (or even billions) of parameters. It is simply impossible. Therefore, the solution is to iteratively adjust parameters and see if the loss improves. This process is called __Gradient Descent__.

$$ \theta = \theta − \eta * \nabla_\theta L(\theta) $$

This process is still highly complex since every node contains the accumulation of all mathematical operations of previous layers and nodes. This is a multivariate problem, which requires the computation of all partial derivatives for every weight.

The deeper the network the more infeasible the process becomes. 


### 1.7. Backpropagation

Backpropagation is an algorithm for calculating the gradient of a loss function with respect to variables of a model.

The algorithm computes the gradients of variables in graph structures in a backward prespective, from the output of the graph toward the input, propagating the error in the predicted output that is used to calculate gradient for each variable.

It relies on the __Chain Rule__ of calculus, a formula to compute the derivatives of functions using related functions whose derivatives are known. Since mathematical operations are sequential, it is possible to compute automatically and efficienty the derivatives of all network parameters.

### 2.4. Training a Multilayer Perceptron with Tensorflow

With all the basics of DNNs training, we'll use the Tensorflow Library to train our first network using the MNIST dataset. The network will have __2 fully connected layers__ and __1 output layer__.

For this purpose, it is necessary to define:
* the weights and biases
* the network structure
* the loss function
* the evaluation metric
* the optimizer
* the optimization method

_Note_: This network is based on: https://github.com/aymericdamien/TensorFlow-Examples/blob/master/tensorflow_v2/notebooks/3_NeuralNetworks/neural_network_raw.ipynb

##### 2.4.1. Defining Parameters

```python
# A random value generator to initialize weights.
random_normal = tf.initializers.RandomNormal()

weights = {
    'h1': tf.Variable(random_normal([28*28, 128])),
    'h2': tf.Variable(random_normal([128, 256])),
    'out': tf.Variable(random_normal([256, 10]))
}
biases = {
    'b1': tf.Variable(tf.zeros([128])),
    'b2': tf.Variable(tf.zeros([256])),
    'out': tf.Variable(tf.zeros([10]))
}
```

##### 2.4.2. Setup the Network

```python
def neural_network(x):
    
    # Hidden fully connected layer with 128 neurons.
    layer_1 = tf.add(tf.matmul(x, weights['h1']), biases['b1'])
    # Apply sigmoid to layer_1 output for non-linearity.
    layer_1 = tf.nn.sigmoid(layer_1)
    
    # Hidden fully connected layer with 256 neurons.
    layer_2 = tf.add(tf.matmul(layer_1, weights['h2']), biases['b2'])
    # Apply sigmoid to layer_2 output for non-linearity.
    layer_2 = tf.nn.sigmoid(layer_2)
    
    # Output fully connected layer with a neuron for each class.
    out_layer = tf.matmul(layer_2, weights['out']) + biases['out']
    
    # Apply softmax to normalize the logits to a probability distribution.
    return tf.nn.softmax(out_layer)
```

##### 2.4.3. Defining the Loss function

```python
# Cross-Entropy loss function.
def cross_entropy(y_pred, y_true):
    # Encode label to a one hot vector.
    y_true = tf.one_hot(y_true, depth=10)
    # Clip prediction values to avoid log(0) error.
    y_pred = tf.clip_by_value(y_pred, 1e-9, 1.)
    # Compute cross-entropy.
    return tf.reduce_mean(-tf.reduce_sum(y_true * tf.math.log(y_pred)))
```

##### 2.4.4. Defining the Metric and Optimizer

```python
# Accuracy metric.
def accuracy(y_pred, y_true):
    # Predicted class is the index of highest score in prediction vector (i.e. argmax).
    correct_prediction = tf.equal(tf.argmax(y_pred, 1), tf.cast(y_true, tf.int64))
    return tf.reduce_mean(tf.cast(correct_prediction, tf.float32), axis=-1)

# Stochastic gradient descent optimizer.
optimizer = tf.optimizers.SGD(0.001)
```

##### 2.4.5. Defining the optimization process

```python
# Optimization process. 
def run_optimization(x, y):
    # Wrap computation inside a GradientTape for automatic differentiation.
    with tf.GradientTape() as g:
        pred = neural_network(x)
        loss = cross_entropy(pred, y)
        
    # Variables to update, i.e. trainable variables.
    trainable_variables = list(weights.values()) + list(biases.values())

    # Compute gradients.
    gradients = g.gradient(loss, trainable_variables)
    
    # Update W and b following gradients.
    optimizer.apply_gradients(zip(gradients, trainable_variables))
```

##### 2.4.6. Train the Network

```python
training_steps = 1500
batch_size = 256

X_train_tf, X_test_tf, X_val_tf = X_train.reshape([-1, 28*28]), X_test.reshape([-1, 28*28]), X_val.reshape([-1, 28*28])
train_data = tf.data.Dataset.from_tensor_slices((X_train_tf, y_train))
train_data = train_data.repeat().shuffle(5000).batch(batch_size).prefetch(1)

# Run training for the given number of steps.
for step, (batch_x, batch_y) in enumerate(train_data.take(training_steps), 1):
    # Run the optimization to update W and b values.
    run_optimization(batch_x, batch_y)
    
    if step % 100 == 0:
        pred = neural_network(batch_x)
        loss = cross_entropy(pred, batch_y)
        acc = accuracy(pred, batch_y)
        
        pred_val = neural_network(X_val_tf)
        val_loss = cross_entropy(pred_val, y_val)
        val_acc = accuracy(pred_val, y_val)
        print("Step: %i | Train Loss: %f, Train Accuracy: %f | Val Loss: %f, Val Accuracy: %f" % (step, loss, acc, val_loss, val_acc))

# Test model on validation set.
pred = neural_network(X_test_tf)
print("Test Accuracy: %f" % accuracy(pred, y_test))
```

### 3.1. Training a Multilayer Perceptron with Keras

The API for NN training is more straightfoward. To train a multilayer perceptron the necessary steps are:

* Defining the model using pre-built layers.
* Call a compile function, where the loss, optimizer and other hyperparameters are defined.
* Train the network with the `.fit` method.

##### 3.1.1. Defining the Multilayer Perceptron

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=[28, 28]),
    tf.keras.layers.Dense(300, activation="relu"),
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(10, activation="softmax")
])
```

```python
model.summary()
```

##### 3.1.2. Defining the Network Compiler

```python
model.compile(loss="categorical_crossentropy",
              optimizer="sgd",
              metrics=["accuracy"])
```

##### 3.1.3. Train the Network

```python
# transform labels into one-hot-encoded vectors
y_train = tf.keras.utils.to_categorical(y_train, num_classes = 10)
y_test = tf.keras.utils.to_categorical(y_test, num_classes = 10)
y_val = tf.keras.utils.to_categorical(y_val, num_classes = 10)

history = model.fit(X_train, y_train, epochs=15, validation_data=(X_val, y_val), batch_size=256)
```

##### 3.1.4. Evaluating Results

__Loss and Accuracy Curves__

```python
import pandas as pd

pd.DataFrame(history.history).plot(figsize=(8, 5))
plt.grid(True)
plt.gca().set_ylim(0, 1) # set the vertical range to [0-1]
plt.show()
```

__Verify the Test Set__

```python
model.evaluate(X_test, y_test)
```

```python
y_test_pred = np.argmax(model.predict(X_test), axis=1)
```

__Visualizing Predictions__

```python
def visualize_predictions(X, y, y_pred, n_images, wrong=False):
    if wrong:
        errors = np.argmax(y_test, axis=1) != y_test_pred
        
        X = X[errors]
        y = y[errors]
        y_pred = y_pred[errors]
        
    # Display image and model prediction.
    for i in range(n_images):
        plt.figure(figsize=(3, 3))
        select_img = np.random.choice(np.arange(len(X)))
        plt.imshow(np.reshape(X[select_img], [28, 28]), cmap='gray')
        plt.axis("off")
        plt.show()
        
        print("Real Label: %i" % np.argmax(y[select_img]))
        print("Model prediction: %i" % y_pred[select_img])
```

### 3.2. Training a Convolution Neural Network with Keras

The MNIST dataset includes images of hand-written numbers. As we discussed, Convolutional Neural Networks offer the possibility of exploring spatial dependencies through filters. Therefore, this problem should benefit from CNNs. Let's try them.

```python
from functools import partial

DefaultConv2D = partial(tf.keras.layers.Conv2D, kernel_size=3, activation='relu', padding="SAME")

model = tf.keras.models.Sequential([
    DefaultConv2D(filters=64, kernel_size=7, input_shape=[28, 28, 1]),
    tf.keras.layers.MaxPooling2D(pool_size=2),
    DefaultConv2D(filters=128),
    #DefaultConv2D(filters=128),
    tf.keras.layers.MaxPooling2D(pool_size=2),
    DefaultConv2D(filters=256),
    #DefaultConv2D(filters=256),
    tf.keras.layers.MaxPooling2D(pool_size=2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(units=128, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(units=64, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(units=10, activation='softmax'),
])
```
